#1. Consider x = [1, 2, 3, 3, 4, 5, 5, 5, 6, 8, 0, 10]

(a) Write a program to find the frequency (number of occurrences) of each element
in x.

(b) Print the maximum and minimum element in x.

(c) Write a program for arranging the elements in x in ascending and descending
order.

````python
# Initialization of list
x = [1, 2, 3, 3, 4, 5, 5, 5, 6, 8, 0, 10]
print(x)
````

**a) Write a program to find the frequency (number of occurrences) of each element in x.**

````python
# Set stores unique values. It removes duplicates. So, By making a set from the given list results a collection of unique elements.
new_set = set(x)
# As new_set has unique elements, we can iterate over set and find the number of occurences of each element in the given list. 
# Following list comprehension will return list of tuples (a,b), where "a" is the element in "x" and "b" is the frequency of the "a" in "x"
frequency_tuple = [(i,x.count(i)) for i in new_set]
print(frequency_tuple)
````

**(b) Print the maximum and minimum element in x.**

````python
# max() function return maximum value in given iterable
print(f"The maximun element in x is {max(x)}")

# min() function reutrn minimum value in given iterable
print(f"The minimun element in x is {min(x)}")
````

````python
# Finding MAX and MIN element without using max() and min() functions

# Initialising mini and maxi variables with first element of the list
maxi=mini=x[0]

for i in x:
  # If an element in x is greater than maxi, then updating the maxi
  if i > maxi:
    maxi = i
  # If an element in x is less than mini, updating the mini
  if i < mini:
    mini = i
print(f"The maximun element in x is {maxi}")
print(f"The minimun element in x is {mini}")
````

**(c) Write a program for arranging the elements in x in ascending and descending order.**

````python
# sorted(iterable, *, key=None, reverse=False)
# sorted function return a copy of sorted list.
# Reverse flag can be used to sort the given list in ascending and descending order
# If reverse = True, the function returns a list sorted in descending order
# If reverse = False, the function returns a list sorted in ascending order. By default reverse flag is set to False

# Ascending order

print(f"Ascending Order :  {sorted(x)}")

# Descending order

print(f"Descending order :  {sorted(x,reverse=True)}")
````

````python
# Function to sort a list.
# If descending = True, then the list will be sorted in  descending order.
# If descending = False, then the list will be sorted in  ascending order.
# Default value of descending is False

def sorting(collection,descending=False):
  # Selection Sort Algorithm is Used
  # In this algorithm, the list is dividing into two parts. i.e., sorted list and unsorted list
  # In every iteration (for ascending order) , the minimum element in the unsorted list is picked and appended to sorted list.
  # Initially the sorted list will be empty.
  
  # In each iteration, the "i"th index will get minimum (or maximum) element from "i+1" index to end of the list
  for i in range(len(collection)-1):
    index = i
    # iterating over unsorted list to find minimum ( or maximum) element
    for j in range(i+1,len(collection)):
      # If descending flag is set to True, we need to sort the list in descending order
      # If we find an element which is greater than the element whose index is in "index", then update the index.
      # At the end of the iteration, index of maximum element in unsorted list will be in "index" variable
      if descending and collection[j] > collection[index]:
        index = j
      # If descending is False, we need to sort in ascending order. 
      # At the end of iteration, index of minimum element of unsorted list will be in "index" variable
      elif not descending and collection[j] < collection[index]:
        index = j
    if index != i:
      # Swapping the elements and getting the minimum (or maximum element at required position)
      collection[i],collection[index] = collection[index],collection[i]
  return collection

# Ascending order

print(f"Ascending Order :  {sorting(x)}")

# Descending order

print(f"Descending order :  {sorting(x,True)}")
````

#5. Consider S=[a,o,h,j,e,j]. Write a program to perform the following task: 

Read each alphabet from S. If the alphabet is a vowel: print ”The alphabet is a vowel”. Construct a string by concatenating the alphabet three times. Else: print ”The alphabet is a consonant”. Construct a string by concatenating the alphabet five times.

````python
# Initialising the list
s = ["a","o","h","j","e","j"]

# List of all vowels
vowels = ["a","e","i","o","u"]

# Iterating over the list "s"
for i in s:
  # If the element of "s" is in "vowels" list, then the element is a vowel
  if i in vowels:
    print("The alphabet is a vowel")
    # Concatenating the character to itself, three times and the resulting string is stored in a variable
    temp_string = i*3
    print(temp_string)
  # If the element of "s" is not in "vowels" list, then the element is a consonant
  else:
    print("The alphabet is a consonant")
    # Concatenating the character to itself, five times and the resulting string is stored in a variable
    temp_string = i*5
    print(temp_string)
````

````python
def print_matrix(mat):
  ''' Prints the given matrix ''' 
  for i in mat:
    for j in i:
      print(j,end=" ")
    print()
````

#8. 

```
      3 -1  4            2 -1  1
A =   1  5  6      B =   3  2 -6
      1  0  2            4 -5  2
```
(a) Using a suitable data structure construct A and B.

(b) Find A + B,

(c) Find A ∗ B

(d) Construct a 3 × 6 matrix by concatenating A and B.

(e) Construct a 6 × 3 matrix by concatenating A and B.


````python
def print_matrix(mat):
  ''' Prints the given matrix ''' 
  for i in mat:
    for j in i:
      print(j,end=" ")
    print()
````

**(a) Using a suitable data structure construct A and B.**

````python
A = [
      [ 3 , -1 , 4 ],
      [ 1 ,  5 , 6 ],
      [ 1 ,  0 , 2 ]
]

B = [
     [ 2 , -1 , 1 ],
     [ 3 ,  2 ,-6 ],
     [ 4 , -5 , 2]
]

print_matrix(A)
print()
print_matrix(B)
````

**(b) Find A + B**

````python
# Matrix addition is possible only if size of both the matrices are same

if len(A) == len(B) and len(A[0]) == len(B[0]):
  # List to store resultant matrix
  sum_matrix = []
  for i in range(len(A)):
    # List to temporarily store rows of resultant matrix
    temp = []
    for j in range(len(A[0])):
      # Adding corresponding elements of A and B & appending the result value to "temp" list , i.e., current row.
      temp.append(A[i][j]+B[i][j])
    # Adding the resultant row to matrix
    sum_matrix.append(temp)
  print_matrix(sum_matrix)
````

**(c) Find A ∗ B**

````python
# Matrix multiplication, AB, is possible only if number of columns of A is equal to number of rows of B
# The c(i,j) of resultant matrix C is obtained by adding the products corresponding elements of row i in A and column j in B. 
if len(A[0]) == len(B):
  # List to store resultant matrix
  product_matrix = []
  # Iterating over rows of A
  for i in range(len(A)):
    # List to temporarily store rows of resultant matrix
    temp = []
    # Iterating over columns of B
    for j in range(len(B[0])):
      # "temp_var" is used to store the value product_matrix[i][j] temporarily
      temp_var = 0
      # product_matrix[i][j] is obtained by adding the products of A[i][k] and  B[k][j], where k varies from 0 to number of rows in B
      for k in range(len(B)):
        temp_var += A[i][k]*B[k][j]
      temp.append(temp_var)
    product_matrix.append(temp)
  print_matrix(product_matrix)
````

**(d) Construct a 3 × 6 matrix by concatenating A and B.**

````python
# 3x6 matrix can be formed by concatenating the rows of two matrices.
new_matrix = [ A[i]+B[i] for i in range(len(A))]
print_matrix(new_matrix)
````

**(e) Construct a 6 × 3 matrix by concatenating A and B.**

````python
# 6x3 matrix can be formed by concatenating the two lists A and B

new_matrix2 = A+B
print_matrix(new_matrix2)
````
