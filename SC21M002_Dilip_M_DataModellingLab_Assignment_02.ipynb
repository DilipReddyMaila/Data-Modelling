#2. (a) Create a matrix using the following information 
First row:(6,4,2), second row:(-5,-3,-3), third row: (2,2,3).

(b) Create b vector where b = (1, 1.0, 2)T.

(c) Find eigenvalues and corresponding eigenvectors of A.

(d) Find norm of b.

```python
# Using Lists to create matrix

matrix_1 = [[6,4,2],[-5,-3,-3],[2,2,3]]

vector_b = [1,1.0,2]

# Importing linalg from numpy
from numpy import linalg

# eig() returns eigen values and corresponding eigen vectors
eigen_values , eigen_vectors = linalg.eig(matrix_1)

# norm() return norm of given vector or matrix
norm_b = linalg.norm(vector_b)

# Printing

print("Eigen Values and vectors of given matrix are :")
for i in range(len(eigen_values)) : print(f"\n{eigen_values[i]}\t\t{list(eigen_vectors[i])}")
print("\nNorm of vector b is : ",norm_b)
```

#5. (a) Generate 1000 random numbers from [0,1) and store the first 500 in vector x and remaining in vector y.
(b) Find z = x ∗(x + y). Plot scatter plot for representing z. Set axis labels and
figure title.

```python
import numpy as np
# Generating 1000 random numbers
random_numbers = np.random.rand(1000)

# Storing first 500 numbers in vector_x
vector_x = np.array(random_numbers[:500])

# Storing remaining numbers in vector y
vector_y = np.array(random_numbers[500:])

# z = x*(x+y)
vector_z = vector_x * ( vector_x + vector_y)

# Using pyplot for plotting the scatter plot
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection="3d")
ax.scatter3D(vector_x,vector_y,vector_z)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Scatter Plot")
# plt.xlabel("x")
# plt.ylabel("y")

# plt.title("Scatter Plot")

plt.show()
```

#6. x=[1,-2,4], y=[-3,2,4].
 Plot the hyperplane z = 2x + 3y + 6. Set label and title

```python
# Using numpy arrays to store x and y
import numpy as np
list_x = np.array([1,-2,4])
list_y = np.array([-3,2,4])

# z = 2x + 3y + 6
list_z = 2*list_x + 3*list_y + 6

# Importing pyplot and Axes3D to plot the hyperplane
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Creating a figure
fig = plt.figure()

# Creating a 3 Dimensional Axes
ax = fig.gca(projection="3d")

# Using trisurf(). It results a triangular surface.
ax.plot_trisurf(list_x,list_y,list_z)

# Labelling
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z = 2x + 3y + 6")

ax.set_title("Hyperplane")

plt.show()
```

```python
# Using numpy arrays to store x and y
import numpy as np
list_x = np.array([1,-2,4])
list_y = np.array([-3,2,4])

# z = 2x + 3y + 6
X,Y = np.meshgrid(list_x,list_y)
Z = 2*X+ 3*Y+ 6

# Importing pyplot and Axes3D to plot the hyperplane
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Creating a figure
fig = plt.figure()

# Creating a 3 Dimensional Axes
ax = fig.gca(projection="3d")

# Using plot_surface() to plot
ax.plot_surface(X,Y,Z)

# Labelling
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z = 2x + 3y + 6")

ax.set_title("Hyperplane")

plt.show()
```

```python
# Plotting the plane using the given equation z = 2x + 3y + 6

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generating numbers from -5 to 5
x = range(-5,6)
y = range(-5,6)

# Using meshgrid() for generating coordinate matrix
# If x = [1,2] , y= [2,3]
# meshgrid(x,y) will return X,Y
# X = [[1,2],[1,2]] Y = [[2,2],[3,3]]
# Which corresponds to (x,y) -> (1,2) , (2,2) , (1,3) , (2,3)

X,Y = np.meshgrid(x,y)

# Finding the Z
Z = 2*X + 3*Y + 6

# Creating a figure
fig = plt.figure()

# Creating a 3 Dimensional Axes
ax = fig.gca(projection="3d")

# Using plot_surface() to plot
ax.plot_surface(X,Y,Z)

# Labelling
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z = 2x + 3y + 6")

ax.set_title("Hyperplane")

plt.show()
```

**(a) f : [−3, 3] →R where f(x) = cosx + 1/( 1 + exp −(2x)).**

```python
import numpy as np
from matplotlib import pyplot as plt

# Generating 100 numbers from -3 to 3
x = np.linspace(-3,3,100)

# Finding f(x)
f = np.cos(x) + 1/(1+np.exp(-2*x))

# Plotting the function
plt.plot(x,f)

#Labelling
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = cosx + 1/( 1 + exp −(2x))")

plt.show()
```

**(b) f : [−16, 10] → R where f(x) = (2 + 2x)/ (5sinx −6). Draw the tangent line to f at x=-6.**

```python
import numpy as np
from matplotlib import pyplot as plt

# Generating 400 numbers from -16 to 10
x = np.linspace(-16,10,400)

# Finding f(x)
f = (2 + 2*x)/(5*np.sin(x)-6)

# For finding the tangent, two neighbourhood point of -6 i.e., (-6 - 1.0e-10) and (-6 + 1.0e-10) are considered. The corresponding f(x) for those two values are calculated
# Now we have two data points (x1,y1) and (x2,y2).
# With those two data points we can form a line

tangent_x = np.array([-6 - 1.0e-10 , -6 + 1.0e-10])
tangent_y = (2 + 2*tangent_x)/(5*np.sin(tangent_x)-6)

# Slope of the tangent line is calculated as (y2-y1)/(x2-x1)
slope = (tangent_y[1]-tangent_y[0])/(tangent_x[1]-tangent_x[0])

# We need datapoints to plot tangent
# 30 points are generated in range -16 to 10
slope_x = np.linspace(-16,10,30)

# Corresponding y values for above datapoints are calculated using the line equation i.e., y = slope * (x -x1) + y1
slope_y = slope*(slope_x-tangent_x[0])+tangent_y[0]


# Plotting the function
plt.plot(x,f,'b-',label="Function")
# Plotting the tangent
plt.plot(slope_x,slope_y,'r-',label="Tangent")

#Labelling
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title(" f(x) = (2 + 2x)/ (5sinx −6)")

plt.legend()
plt.show()
```

**(c) f : [0, 2] ×[0, 1] → R where f(x) = ‖x‖2**

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x1 = [0,2]
x2 = [0,1]

X1,X2 = np.meshgrid(x1,x2)

# X1 will be [[0 2] , [0 2]]
# X2 will be [[0 0] , [1 1]]

f = X1**2 + X2**2
# f = [[0 4] . [1 5]]

# X will be { (0,0) , (2,0) , (0,1) , (2,1) }
# f will be { 0 , 4 , 1 ,5}

fig = plt.figure()

ax = fig.gca(projection="3d")

ax.plot_surface(X1,X2,f)

ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("||X||^2")

ax.set_title("f(x)=||x||^2")

plt.show()
```
