# 1. Write a function that could perform the following tasks: 

(a) Generate 1000 data points that satisfy -8x -3y +2z -11 =0.
 
(b) Create a folder in the current directory
and store the date generated in text as well as CSV format.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
import os
import csv
def func1():
  x = np.linspace(-10,10,1000)
  y = np.linspace(-10,10,1000)
  z = (8*x+3*y+11)/2

  os.mkdir(os.path.join(os.getcwd(),"problem_1"))

  with open(os.path.join(os.getcwd(),"problem_1","p1.txt"),"w") as f1 , open(os.path.join(os.getcwd(),"problem_1","p1.csv"),"w") as f2:
    f1.write("x , y , z\n")
    writer = csv.writer(f2)
    writer.writerow(["x","y","z"])
    for i in range(len(x)):
      writer.writerow([x[i],y[i],z[i]])
      f1.write(f"{x[i]},{y[i]},{z[i]}\n")
```

```python
func1()
```

#4. Plot

(a) An undirected graph with vertices: {v 1 , v 2 , v 3 , v 4 } having edges
(v 1 , v 2 ), (v 2 , v 3 ), (v 3 , v 4 ). Mark edges as E 1 , E 2 and E 3 .

(b) A directed graph using the above information.

```python
import networkx as nx 

G = nx.Graph()

G.add_nodes_from(["v1","v2","v3","v4"])
G.add_edges_from( [("v1","v2"),("v2","v3"),("v3","v4")] )

pos = nx.shell_layout(G)

nx.draw_networkx_edge_labels(G,pos,edge_labels={
    ("v1","v2"):"E1",
    ("v2","v3"):"E2",
    ("v3","v4"):"E3"
})
nx.draw_networkx(G,pos,with_labels=True)
```

```python
import networkx as nx 

G = nx.DiGraph()

G.add_nodes_from(["v1","v2","v3","v4"])
G.add_edges_from( [("v1","v2"),("v2","v3"),("v3","v4")] )

pos = nx.shell_layout(G)

nx.draw_networkx_edge_labels(G,pos,edge_labels={
    ("v1","v2"):"E1",
    ("v2","v3"):"E2",
    ("v3","v4"):"E3"
})
nx.draw_networkx(G,pos,with_labels=True)
```

# 5. Write a function that could perform the following tasks: Ask the user to enter a integer n.

(a) Display all the even integers less than or equal to n.

(b) Display all the multiples of 5 less than or equal to n.

(c) Display the first n terms of the exp(x) series.

(d) Display the first n terms of the Fibonacci series.

```python
def func5():
  n = int(input("Enter a number"))
  even = []
  multiple = []
  fib = np.zeros(n)
  fact = np.zeros(n)
  for i in range(n+1):
    if i%2 == 0 : even.append(i) 
    if i%5 == 0 : multiple.append(i)
    if i == 0:
      fact[0] = 1
      fib[0] = 1
    elif i ==1:
      fact[1] = 1
      fib[1] = 1
    elif i < n:
      fact[i] = fact[i-1] * i 
      fib[i] = fib[i-1] + fib[i-2]
  print(f"Even integers less than or equal to {n} are {even}\n")
  print(f"multiples of 5 less than or equal to {n} are {multiple}\n")
  print(f"first {n} terms of the exp(1) series are {1/fact}\n")
  print(f"first {n} terms of the Fibonacci series are {fib}\n")
```

# 6. Write a function that could perform the following tasks: Ask the user to enter n English words. Sort the words in alphabetic (dictionary) order.

```python
def insertion_sort(x):
  for i in range(len(x)-1):
    min = i
    for j in range(i,len(x)):
      if x[j] < x[min]:
        min = j
    x[i],x[min] = x[min],x[i]
  return x
```

```python
def func6():
  n = int(input("Enter n"))
  input_words = [input() for _ in range(n)]
  # input_words.sort()
  print(insertion_sort(input_words))
```

```python
func6()
```
