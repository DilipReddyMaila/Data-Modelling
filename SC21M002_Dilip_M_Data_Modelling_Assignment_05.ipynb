```python
def print_folder(path_name,call,print_str):
  
  print_str += (("#"+" --"*(call)+">"+path_name.split("/")[-1]+"\n"))
  for i in os.listdir(path_name):
    if not os.path.isdir(os.path.join(path_name,i)):
      print_str += (str("."+" --"*(call+1)+">"+i+"\n"))
    else:
      print_str += (print_folder(os.path.join(path_name,i),call+1,""))
  return print_str

print(program_3())
```

#4. Write a function for the following task:
(a) Store the following items in a deque data structure: Apple, Orange, Lemon, 20,
40.

(b) Add 5 in the beginning of deque.

(c) Add Mango at the end of deque.

(d) Reverse the order of deque.

```python
def program_4():
  data = collections.deque(["Apple","Orange","Lemon",20,40])

  data.appendleft(5)

  data.append("Mango") 
  data.reverse()
  
  return list(data)

program_4()
```

# 5. Write a function to print 

(a) path of the current directory 

(b) subdirectories in a given directory.

```python
 
def print_subdirs(path):
  print_str = ""
  sub_dir = os.listdir(path)
  print_str += ("\nSubdirectories in a given directory are :\n")
  for i in sub_dir:
    if os.path.isdir(os.path.join(path,i)):
      print_str += ("# "+str(i)+"\n")
  return print_str

def program_5():
  path = os.getcwd()
  print_str = ""
  print_str += ("Path of the current directory is :"+str(path)+"\n")
  print_str += print_subdirs(path)
  return print_str
  
print(program_5())
```

# 7. Write a program for the following tasks:
(a) Connect a machine to a given host and port.

(b) Read any available data from the socket and then exit.

```python
 

def program_7():

  host = "8.8.8.8"

  port = 443

  conn = socket.socket()

  conn.connect((host,port))

  data = conn.recv(1024)

  conn.close()

  return data 

print(program_7())
```

```python
# (a) Ask the user to enter a name for creating a file. Using that name create a file in the current directory.


file_name = input("Enter file name")

f = open(os.path.join(os.getcwd(),file_name),"w")


```

```python

# (b) Ask the user to enter the details for storing in the file. 
data = ""
inp = input("Enter details to store in file, enter '.' after completion of all lines\n")
while(inp != "."):
  data+=inp
  inp = input()

print(data)
```

```python
# opening the file for reading
f = open(os.path.join(os.getcwd(),file_name))
lines = f.readlines()
```

```python
f.close()
```

```python
# (c) Print the number of words in the file.

words = 0
for i in lines:
  words += len(i.split(" "))

print(words)
```

```python
# (d) Write the program using try-except syntax: Ask the user to enter an integer n.
# Read the n th line from the file. Raise an error if line is not found.

n = int(input("Enter line number to read\n"))
try:
  print(lines[n+1])
except:
  print("Entered line number doesn't exist\n")
```

```python
# (e) Ask the user to enter a digit n. Print the first n lines from the file. Print the
# last n lines of the file. Raise an error if lines are not found.

n = int(input())
try:
  print(f"First {n} lines \n")
  for i in lines[:n]:
    print(i)
  print(f"\nLast {n} lines\n")
  for i in lines[len(lines)-n:]:
    print(i)
except:
  print("Sufficient lines are not found")
```

```python
# (f) Print the longest word in the file.
# (h) Count the frequency of each word in the file.
frequency = {}
longest = ""
for i in lines:
  for j in i.split(" "):
    j = j.strip("\n")
    if len(j) > len(longest):
      longest = j
    try:
      frequency[j] += 1
    except:
      frequency[j] = 1

print("Longest word :",longest)
print("\nFrequency of each word:\n",frequency)
```

```python
# (g) Print all the lines in the file.

for i in lines:
  print(i)
```

```python
# Copy the contents of the file to another file.
with open(os.path.join(os.getcwd(),file_name)) as f, open(os.path.join(os.getcwd(),"new"),"w") as f1:
  data = f.readlines()
  for i in data:
    f1.write(i)
```
