#1. Plot normal distribution with mean 0 and standard deviation 1.

```python

import numpy as np
from matplotlib import pyplot as plt 
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-4,4,100)
y = norm.pdf(x,0,1)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Normal Distribution")
plt.show()
```

```python
# if mean = 0 and standard deviation = 1
# Then normal distribution is f(x) = exp(-0.5*(x^2))/(sqrt(2*phi))

x = np.linspace(-4,4,100)
y = np.exp(-0.5*(x**2)) / np.sqrt(2*np.pi)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Normal Distribution")
plt.show()
```

#2. Plot exp(x), exp(−x), logx(x > 0) in the same figure by giving different shades to each plot.

```python
x = np.linspace(0.001,2,100)
y1 = np.exp(x)
y2 = np.exp(-x)
y3 = np.log(x)

plt.plot(x,y1,'r-',x,y2,"b-",x,y3,"y-")
plt.legend(["exp(x)","exp(−x)","logx"])
plt.show()
```

```python
# (a) Generate 200 data points that satisfy the equation 3x + 4y − 3 = 0. Plot the hyperplane and mark the selected points.

x = np.linspace(-5,5,200)

y = (3-3*x)/4

plt.plot(x,y,"ro-")
plt.show()
```

```python
# (b) Generate 1000 data points that satisfy the equation 5x − 2y − 6z = 0. Plot the hyperplane and mark the selected points.

x = np.linspace(0,10,1000)
y = np.linspace(0,60,1000)
X,Y = np.meshgrid(x,y)

z = (5*X - 2*Y)/6
ax = plt.figure(figsize=[15,15]).gca(projection="3d")
ax.plot_surface(X,Y,z,alpha=0.5)
ax.scatter3D(X,Y,z,c="r")

plt.show()
```

#5. Write a function module for the following tasks: 
(a) To determine the probability
mass function of Bernoulli and Binomial distribution. 

(b) To determine the probability density function of Normal distribution.

```python
# (a) To determine the probability mass function of Bernoulli and Binomial distribution.

def pmf_bernoulli(x,p):
  x = np.array(x)
  return np.power(p,x) * np.power(1-p,1-x)

def ncr(n,r):
  num = 1
  den = 1
  for i in range(r):
    den = den*(i+1)
    num = num* (n-i)
  return num/den

def pmf_binomial(x,p,n):
  x = np.array(x)
  return np.array(list(map(ncr,n*np.ones(len(x)),x))) * np.power(p,x) * np.power(1-p,n-x)
```

```python
# (b) To determine the probability density function of Normal distribution.

def pdf_normal(x,u,sd):
  x = np.array(x)
  return np.exp(-0.5* np.power( ((x-u)/sd) ,2) ) / (sd*np.sqrt(2*np.pi))
```

#6. Draw a figure that consists of the following subfigures: 
(a) Circle 1: x 2 + y 2 = 3

(b) Circle 2: x 2 + y 2 = 2 

(c) Red colour between Circle 1 and Circle 2 

(d) Green colour inside Circle 2.

```python

circle2 = plt.Circle((0,0),radius=np.sqrt(2),color="g",fill=True)
circle1 = plt.Circle((0,0),radius=np.sqrt(3),color="r",fill=True)
ax = plt.figure(figsize=(10,10)).gca()
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.add_patch(circle1)
ax.add_patch(circle2)
plt.show()
```

# 7. Draw a figure that consists of the following subfigures: 

(a) Line 1: 6x − 9y + 2 = 0

(b) Line 2: 6x−9y+2 = 1

(c) Line 3: 6x−9y+2 = −1

(d) A segment that connects
6x − 9y + 2 = 1 and 6x − 9y + 2 = −1. 

(e) Blue colour above 6x − 9y + 2 = 1
and yellow colour below 6x − 9y + 2 = −1.

```python
ax = plt.figure(figsize=(10,10)).gca()

x = np.linspace(-0.5,0.5,20)
y1 = (6*x + 2)/9
y2 = (6*x + 1)/9
y3 = (6*x + 3)/9

X = [0,0]
Y = [1/9,3/9]
ax.plot(x,y1,x,y2,x,y3,X,Y)
ax.legend(["Line1","Line2","Line3","Segement joining 2 and 3"])
ax.fill_between(x,y2,y1,where=(y1>y2),color="b",alpha=0.2)
ax.fill_between(x,y3,y1,where=(y3>y1),color="y",alpha=0.2)
plt.show()
```
