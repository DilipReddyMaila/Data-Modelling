# Using Python object oriented programming techniques create a Library Database Management System that has the following functionalities:

 (i) store the details of members 
 
 (ii) store the details of books 
 
 (iii) rules of the library 
 
 (iv) details of books borrowed by each user 

(v) fine calculation 

(vi) searching facilities (book/member details)

```python
# Class for Book
class Book:

  def __init__(self,id,name,price,author,category,isbn):
    self.id = id
    self.name = name
    self.price = price
    self.author = author
    self.category = category
    self.isbn = isbn


  def details(self,name=False):
    if name:
      print(self.name)
    else:
      print(f"Book ID : {self.id} , Book Name :{self.name} , Price : {self.price} , Author : {self.author} , Category : {self.category} , ISBN : {self.isbn}")

# Class for Members
class Member:

  def __init__(self,id,name,login_name,password):
    self.id = id
    self.name = name
    self.login_name = login_name
    self.password = password

  def details(self):
    print(f"Member ID : {self.id} , Member Name : {self.name} , Login Name : {self.login_name}")

# Class for Member and Book mapping
class MemberBookMap:

  def __init__(self,id,member,book,returned=False):
    self.id = id
    self.member = member
    self.book = book
    self.returned = returned
  
  def details(self) :
    print(f"Map ID : {self.id} ,Member : {self.member.name} , Book : {self.book.name} , Returned : { self.returned }")

# Library Management System    
class LDMS:

  def __init__(self):
    self.members = []
    self.books = []
    self.mem_id = 1
    self.book_id = 1
    self.rules = []
    self.mem_bk_map = []
    self.mem_bk_id = 1

  def create_member(self,name,login_name,password):
    mem = Member(id=self.mem_id,name=name,login_name=login_name,password=password)
    for i in self.members:
      if i.login_name == mem.login_name:
        print("Login Name already exists")
        break
    else:
        self.members.append(mem)
        self.mem_id += 1
        print("Member Created")

  def create_book(self,name,price,author,category,isbn):
    bk = Book(id=self.book_id,name=name,price=price,author=author,category=category,isbn=isbn)
    for i in self.books:
      if i.isbn == bk.isbn:
        print("Book with given ISBN number already exists")
        break
    else:
        self.books.append(bk)
        self.book_id += 1
        print("Book Created")

  def book_search(self,s_by,s_with,print=True):
    result_books = []
    for i in self.books:
      if s_by == "id":
        if i.id == s_with:
          result_books.append(i)
      elif s_by == "name":
        if i.name == s_with:
          result_books.append(i)
      elif s_by == "author":
        if i.author == s_with:
          result_books.append(i)
      elif s_by == "isbn":
        if i.isbn == s_with:
          result_books.append(i)
      elif s_by == "category":
        if i.category == s_with:
          result_books.append(i)
      elif s_by == "price":
        if i.price == s_with:
          result_books.append(i)

    if print:
      for j in result_books:
        j.details()
    else:
      return result_books

  def member_search(self,s_by,s_with,print=True):
    result_members = []
    for i in self.members:
      if s_by == "id":
        if i.id == s_with:
          result_members.append(i)
      elif s_by == "name":
        if i.name == s_with:
          result_members.append(i)
      elif s_by == "login_name":
        if i.login_name == s_with:
          result_members.append(i)
    if print:
      for j in result_members:
        j.details()
    else:
      return result_members

  def add_rule(self,rule):
    self.rules.append(rule)
    print("Rule Added")

  def print_rules(self):
    for i in range(len(self.rules)):
      print(f"{i+1} > {self.rules[i]}\n")
  
  def borrow_book(self,member_id,book_id):
    member = self.member_search(s_by="id",s_with=member_id,print=False)[0]
    book = self.book_search(s_by="id",s_with=book_id,print=False)[0]
    mbm = MemberBookMap(id=self.mem_bk_id,member=member,book=book)
    self.mem_bk_map.append(mbm)
    self.mem_bk_id += 1
    print("Borrowed Book")

  def mem_book_search(self,member_id):
    borrowed_books = []
    for i in self.mem_bk_map:
      if i.member.id == member_id :
        borrowed_books.append(i)
    for j in borrowed_books:
      j.details()
  
  def fine_calc(self,member_id):
    fine = 0
    for i in self.mem_bk_map:
      if i.member.id == member_id:
        if i.returned == False:
          fine += float(i.book.price)
    print(f"Member ID: {member_id} , Fine : {fine}")

  def return_book(self,mbm_id):
    for i in self.mem_bk_map:
      if i.id == mbm_id:
        i.returned = True
        print("Returned")
        break
    else:
      print("Enter correct member_book_map id")
    


  
```
